#Project Data Siswa
Sampel 1000 data Siswa, beragam memiliki 10 kolom

Usia : 16 - 18
Gender : L/P
Jurusan : IPA, IPS, BI
Asal Sekolah : SMP 1, SMP 2, SMP 3
Semester : 1, 2
Kelas : 10, 11, 12
Mata Pelajaran : Matematika, Bahasa Indonesia, Bahasa Inggris
Absen : Hadir, Izin, Sakit, Alfa
KKM  : 75, 80, 90
Kategori : Beasiswa, NonBeasiswa

```python
import numpy as np
import pandas as pd

n_data = 1000

usia = np.random.randint(18,23,size=n_data)
gender = np.random.choice(['L','P'],size=n_data)
jurusan = np.random.choice(['IPA','IPS','BI'],size=n_data)
asalSekolah = np.random.choice(['SMP1','SMP2','SMP3'],size=n_data)
semester = np.random.randint(1,2,size=n_data)
kelas = np.random.choice(['10','11','12'],size=n_data)
mapel = np.random.choice(['MTK','Bahasa Indonesia','Bahasa Inggris'],size=n_data)
absen = np.random.choice(['Hadir','Izin','Sakit','Alfa'],size=n_data)
kkm = np.random.randint(75,80,size=n_data)
kategori = np.random.choice(['Beasiswa','Non Beasiswa'],size=n_data)

df = pd.DataFrame({
    'Usia' : usia,
    'Gender' : gender,
    'Jurusan' : jurusan,
    'Asal Sekolah' : asalSekolah,
    'Semester' : semester,
    'Kelas' : kelas,
    'Mata Pelajaran' : mapel,
    'Absen' : absen,
    'kkm' : kkm,
    'Kategori' : kategori
})

print(df.info)
print(f"Usia siswa termuda: {df.Usia.min()}")
print(f"Usia siswa tertua: {df.Usia.max()}")
jmlL = df[df['Gender']=='L']
jmlP = df[df['Gender']=='P']
print(f"Jumlah mahasiswa laki-laki: {jmlL.size/10}")
print(f"Jumlah mahasiswa perempuan: {jmlP.size/10}")
jmlIPA = df[df['Jurusan']=='IPA']
jmlIPS = df[df['Jurusan']=='IPS']
jmlBI = df[df['Jurusan']=='BI']
print(f"Jumlah siswa jurusan IPA: {jmlIPA.size/10}")
print(f"Jumlah siswa jurusan IPS: {jmlIPS.size/10}")
print(f"Jumlah siswa jurusan BI: {jmlBI.size/10}")
jmlSMP1 = df[df['Asal Sekolah']=='SMP1']
jmlSMP2 = df[df['Asal Sekolah']=='SMP2']
jmlSMP3 = df[df['Asal Sekolah']=='SMP3']
print(f"Jumlah asal sekolah SMP1: {jmlSMP1.size/10}")
print(f"Jumlah asal sekolah SMP2: {jmlSMP2.size/10}")
print(f"Jumlah asal sekolah SMP3: {jmlSMP3.size/10}")
smstrDua = df[df['Semester']>=2]
print(f"Jumlah mahasiswa semester Dua: {smstrDua.size}")
sepuluh = df[df['Kelas']=='10']
sebelas = df[df['Kelas']=='11']
duabelas= df[df['Kelas']=='12']
print(f"Jumlah siswa kelas reguler: {sepuluh.size/10}")
print(f"Jumlah siswa kelas sebelas: {sebelas.size/10}")
print(f"Jumlah siswa kelas duabelas: {duabelas.size/10}")
mtk = df[df['Mata Pelajaran']=='MTK']
print(f"Jumlah siswa yang belajar matematika: {mtk.size/10}")
absen = df[df['Absen']=='Hadir']
print(f"Jumlah absen yang hadir dikelas: {absen.size/10}")
print(f"Rata-rata kkm: {df.kkm.mean()}")
bea = df[df['Kategori']=='Beasiswa']
non = df[df['Kategori']=='Non Beasiswa']
print(f"Jumlah siswa beasiswa: {bea.size/10}")
print(f"Jumlah siswa non-beasiswa: {non.size/10}")

df.to_csv('dataSiswa.CSV')
```
